```python
#driver configuration
opciones=Options()

opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
opciones.add_experimental_option('useAutomationExtension', False)
opciones.headless=False    # si True, no aperece la ventana (headless=no visible)
opciones.add_argument('--start-maximized')         # comienza maximizado
#opciones.add_argument('user-data-dir=selenium')    # mantiene las cookies
#opciones.add_extension('driver_folder/adblock.crx')       # adblocker
opciones.add_argument('--incognito')
```

```python
#extraemos tabla que nos proporcina el coste por m2.
def extraer(url):
    driver=webdriver.Chrome(PATH)
    bankinter='https://www.bankinter.com/blog/finanzas-personales/precio-vivienda-ciudades'
    driver.get(bankinter)
    time.sleep(2)
    #cookies
    aceptar=driver.find_element(By.XPATH, '//*[@id="modal-cookies"]/div[2]/div/div/div/div/div/a')
    aceptar.click()
    time.sleep(2)

    tabla=driver.find_element(By.XPATH, '//*[@id="zonaContenidos"]/section/div/div/div[6]/table')


    tabla=driver.find_element(By.TAG_NAME, 'tbody')

    filas=tabla.find_elements(By.TAG_NAME, 'tr')
    data=[]

    for f in filas:

        elementos=f.find_elements(By.TAG_NAME, 'td') 
        

        tmp=[]

        for e in elementos:

            tmp.append(e.text)

        tmp.append(url.split('/')[-1])  
        data.append(tmp)


    cabeceras=driver.find_element(By.TAG_NAME, 'span')

    cabeceras=[c.text for c in cabeceras.find_elements(By.TAG_NAME, 'strong')]


    driver.quit()

    return pd.DataFrame(data, columns=cabeceras)


```

```python
import regex as re
```

```python
subasta1= pd.read_csv('../data/subastas.csv',sep=',', encoding='utf-8')
```

```python
subasta1.Fecha_de_inicio[1]
```

```python

def limpiar_date(Fecha_de_inicio): 
    Fecha_de_inicio = re.findall('^\d{2}\-\d+\-\d+', Fecha_de_inicio)
    Fecha_de_inicio = ''.join(Fecha_de_inicio)
    return Fecha_de_inicio
    
```

```python
subasta1['Fecha_de_conclusion']=subasta1['Fecha_de_conclusion'].apply(limpiar_date)
```

```python
subasta1.to_csv('../data/subastas.csv', index=False, sep=',') 
```
